```python
# Goal is to web scrape from AccuWeather to create a dataframe with the history of weather for a specific city over time

# link: http://www.accuweather.com/en/us/new-york-ny/10007/june-weather/349727?monyr=6/1/2015&view=table

# imports
import requests
import pandas as pd 
import numpy as np 
from bs4 import BeautifulSoup
from matplotlib import pyplot
import re
from datetime import datetime
```

```python
# set up link
link_base = 'http://www.accuweather.com/en/us/new-york-ny/10007/june-weather/349727?monyr='
link_month = '6'
link_day = '/1/'
link_year = '2015'
link_end = '&view=table'
full_link = link_base + link_month + link_day + link_year + link_end
```

```python
# scrape weather row by row from website

weather = {'DayofWeek':None,'Date':None,'High':None,'Low':None,'precip (in.)':None,'Snow (in.)':None,'Avg. Hi':None,'Avg. Lo':None}
weather_data = []
r = requests.get(full_link)
b = BeautifulSoup(r.text)
weather_detail = b.findAll(name = 'table', attrs = {'class':'stats'})
weather_text = weather_detail[0].text.encode('ascii','ignore').split('\n')
place = 15
for i in range(31):
    try:
        weather['DayofWeek'] = weather_text[place][0:3]
        weather['Date'] = weather_text[place][3:]
        weather['High'] = int(weather_text[place + 1])
        weather['Low'] = int(weather_text[place + 2])
        weather['precip (in.)'] = float(re.findall(r'^[0-9]*[.,]?[0-9]*',weather_text[place + 3])[0])
        weather['Snow (in.)'] = float(re.findall(r'^[0-9]*[.,]?[0-9]*',weather_text[place + 4])[0])
        weather['Avg. Hi'] = int(weather_text[place + 6])
        weather['Avg. Lo'] = int(weather_text[place + 7])
        weather_data.append(weather.copy())
        place += 10
    except:
        pass

# create dataframe
cols = ['DayofWeek','Date','High','Low','precip (in.)','Snow (in.)','Avg. Hi','Avg. Lo']
nyc_weather_df = pd.DataFrame(weather_data, columns = cols)
```
