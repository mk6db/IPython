```python
# Matplotlib examples
# http://jeffskinnerbox.me/notebooks/matplotlib-2d-and-3d-plotting-in-ipython.html

# import matplotlib & numpy
import matplotlib.pyplot as plt
import numpy as np

# another option is to enter "from pylab import *"

# magic function to have graphs show up in ipython notebook
%matplotlib inline
```

```python
# simple matplot figure
x = np.linspace(0, 5, 10)
y = x ** 2

plt.figure()
plt.plot(x, y, 'r') # 'r' changes the color of the line
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```

```python
# add subplots and change the line format
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```

```python
# store a reference to the newly create figure
# (rather than creating a new global reference for a figure)

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
# with a specific reference, we have full control over the plot axis
# we can also add more than one axis to the figure

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

# insert
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title')
```

```python
# if we don't care to explicit about where our plot axes are placed
# in the figure canvas, then we can use one of the many axis
# layout managers in matplotlib. A good option is subplots

fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
# creating two plots at once using subplots
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
```

```python
# we can deal with overlapping graph by using the
# "fig.tight_layout" method, which automatically adjusts
# the positions of the axes on the figure canvas so that there
# is no overlapping content

fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
fig.tight_layout()
```

```python
# matplotlib allows the aspect ratio, DPI and figure size to be
# specified with the "Figure" object is created, using the 
# figsize and dpi keyword arguments. figsize is a tuple with
# width and height of the figure in inches, and dpi is the
# dot-per-inch (pixel per inch)

# creating a figure with size 800 by 400 pixels
fig = plt.figure(figsize=(8,4), dpi = 100)

# the same arguments can also be passed to layout managers
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
# to save a figure to a file we can use the savefig method
fig.savefig('filename.png')

# we can also optionally specify the DPI and choose output formats
fig.savefig('filename.png', dpi = 200)
fig.savefig('filename.svg')
```

```python
# creating a legend
fig, ax = plt.subplots()

ax.plot(x, x**2, label = 'curve1')
ax.plot(x, x**3, label = 'curve2')
ax.legend()
```

```python
# LaTeX is a word processor and a document markup language
# it's best to use "raw" text strings with LaTeX
# to avoid backslash issues

fig, ax = plt.subplots()

ax.plot(x, x**2, label = r'$y = \alpha^2$')
ax.plot(x, x**3, label = r'$y = \alpha^3$')
ax.set_xlabel(r'$\alpha$', fontsize = 18)
ax.set_ylabel(r'$y$', fontsize = 18)
ax.set_title('title')
ax.legend(loc = 2) # uppper left corner
```

```python
# we can also change the global font size and font family
# which applies to all text elements in a figure
# (tick labels, axis labels and titles, legends, etc.)

# update the matplotlib configuration parameters:
plt.rcParams.update({'font.size':18, 'font.family':'serif'})
```

```python
# create a plot with the updated configuration parameters
fig, ax = plt.subplots()

ax.plot(x, x**2, label = r'$y = \alpha^2$')
ax.plot(x, x**3, label = r'$y = \alpha^3$')
ax.set_xlabel(r'$\alpha$', fontsize = 18)
ax.set_ylabel(r'$y$', fontsize = 18)
ax.set_title('title')
ax.legend(loc = 2) # uppper left corner
```

```python
# restore to original configuration parameters
plt.rcParams.update({'font.size': 12, 'font.family': 'sans'})
```

```python
# MATLAB style line color and style
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

```python
# in matplotlib we can also define colors by their name or RGB
# hex codes, and optionally provide an alpha value, using the 
# color and alpha keyword arguments

fig, ax = plt.subplots()

ax.plot(x, x+1, color="red", alpha=0.5) # half-transparant red
ax.plot(x, x+2, color="#1155dd")        # RGB hex code for a bluish color
ax.plot(x, x+3, color="#15cc55")        # RGB hex code for a greenish color
```

```python
# line and marker styles
# to change the line width we can use the linewidth or lw keyword argument
# and line style can be selected using linestyre or ls
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color = 'blue', linewidth = 0.25)
ax.plot(x, x+2, color = 'blue', linewidth = 0.50)
ax.plot(x, x+3, color = 'blue', linewidth = 1.00)
ax.plot(x, x+4, color = 'blue', linewidth = 2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="red", lw=2, linestyle='-')
ax.plot(x, x+6, color="red", lw=2, ls='-.')
ax.plot(x, x+7, color="red", lw=2, ls=':')
ax.plot(x, x+7.5, color="red", lw=2, ls='steps')

# custom dash
line, = ax.plot(x, x+8, color = 'black', lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="green", lw=2, ls='*', marker='+')
ax.plot(x, x+10, color="green", lw=2, ls='*', marker='o')
ax.plot(x, x+11, color="green", lw=2, ls='*', marker='s')
ax.plot(x, x+12, color="green", lw=2, ls='*', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=2, markeredgecolor="blue");
```

```python
# plot range

fig, axes = plt.subplots(1, 3, figsize = (12,4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title('default axes ranges')

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title('tight axes')

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title('custom axes range')
```

```python
# place of ticks and custom tick labels

fig, ax = plt.subplots(figsize = (10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$',r'$\beta$',r'$\gamma$',r'$\delta$',
                   r'$\epsilon$'], fontsize = 18)

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
# use LaTeX formatted labels
ax.set_yticklabels(['$%.1f$' % y for y in yticks], fontsize = 18)
```

```python
# axis grid

fig, axes = plt.subplots(1, 2, figsize=(10,3))

# default grid appearance
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# custom grid appearance
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color = 'b', alpha = 0.5, linestyle = 'dashed', linewidth = 0.5)
```

```python
# we can also change the properties of the axis spines
fig, ax = plt.subplots(figsize=(6,2))

ax.spines['bottom'].set_color('blue')
ax.spines['top'].set_color('blue')

ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)

# turn off axis spine to the right
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()
```

```python
# twin axes
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color = 'blue')
ax1.set_ylabel(r'area $(m^2)$', fontsize=18, color='blue')
for label in ax1.get_yticklabels():
    label.set_color('blue')

ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color='red')
ax2.set_ylabel(r'volume $(m^3)$', fontsize=18, color='red')
for label in ax2.get_yticklabels():
    label.set_color('red')
```

```python
# axes that cross 0 on the x and y axes
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0)) # set position of y spine to y = 0

xx = np.linspace(-.75, 1., 100)
ax.plot(xx, xx**3)
```

```python
# other 2D plot styles
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))

axes[1].step(n, n**2, lw=2)

axes[2].bar(n, n**2, align='center', width=0.5, alpha=0.5)

axes[3].fill_between(x, x**2, x**3, color = 'green', alpha=0.5)
```

```python
# colormap and contour figures
alpha = 0.7
phi_ext = 2 * math.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p)*np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)

phi_m = np.linspace(0, 2*math.pi, 100)
phi_p = np.linspace(0, 2*math.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
```

```python
# list of pre-defined colormaps
# http://www.scipy.org/Cookbook/Matplotlib/Show_colormaps
```

```python
# pcolor
import matplotlib.cm as cm

fig, ax = plt.subplots()

p = ax.pcolor(X/(2*math.pi), Y/(2*math.pi), Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p)
```

```python
fix, ax = plt.subplots()

cnt = plt.contour(Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0,1,0,1])
```

```python
# 3D figures
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
# surface plots
fig = plt.figure(figsize=(14,6))

# 'ax' is a 3D-aware axis instance, because of projection = '3d'
# keyword argument to add_subplot
ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 2, 2, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                   linewidth=0, antialiased=False)
cb = fig.colorbar(p)
```

```python
# wire-frame plot
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')

p = ax.plot_wireframe(X, Y, Z, rstride=4, cstride=4)
```

```python
# coutour plots with projections
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1,1,1, projection='3d')

ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
cset = ax.contour(X, Y, Z, zdir='z', offset=-math.pi, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=-math.pi, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=3*math.pi, cmap=cm.coolwarm)

ax.set_xlim3d(-math.pi, 2*math.pi)
ax.set_ylim3d(0, 3*math.pi)
ax.set_zlim3d(-math.pi, 2*math.pi)
```
