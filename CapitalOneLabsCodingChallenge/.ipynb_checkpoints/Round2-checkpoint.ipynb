```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# read data into one dataframe
folder_link = 'C:/Users/Michael/Desktop/C1DSChallenge/'
file_link = 'AK.txt'
df = pd.read_csv(folder_link + file_link,sep=',',header=None,names=['State','Sex','BirthYear','Name','Count'])

# create list of file names to loop through
file_list = ['AK','AL','AR','AZ','CA','CO','CT','DC','DE','FL','GA','HI','IA','ID','IL','IN','KS','KY','LA','MA','MD','ME','MI','MN','MO','MS','MT','NC','ND','NE','NH','NJ','NM','NV','NY','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VA','VT','WA','WI','WV','WY']

for i in file_list[1:]:
    df = df.append(pd.read_csv(folder_link + i + '.txt',sep=',',header=None,names=['State','Sex','BirthYear','Name','Count']))

df.head()
```

```python
# 4. Of the names represented in the data, find the name that has had the largest
# percentage increase in popularity since 1980. Largest decrease?

# create a DF with only Name, BirthYear, Count

df_pop = df.copy()
df_pop = df_pop[(df_pop.BirthYear == 1980) | (df_pop.BirthYear == 2014)].reset_index()
df_pop['Count_1980'] = df_pop[df_pop.BirthYear == 1980].Count
df_pop['Count_2014'] = df_pop[df_pop.BirthYear == 2014].Count
df_pop.fillna(0,inplace=True)
df_pop = df_pop.groupby(['Name'])['Count_1980','Count_2014'].sum().reset_index()

df_pop_change = df_pop.copy()
df_pop_change = df_pop_change[df_pop.Count_1980 != 0]
df_pop_change['Change'] = df_pop_change.Count_2014 / df_pop_change.Count_1980 - 1
df_pop_change.sort('Change',ascending=False,inplace = True)
highest_per_change = df_pop_change[df_pop_change.Change == df_pop_change.Change.max()]
print 'The name(s) with the highest percentage increase in popularity was:',' '.join(list(highest_per_change.Name))

df_pop_change.sort('Change',ascending=True,inplace = True)
lowest_per_change = df_pop_change[df_pop_change.Change == df_pop_change.Change.min()]
print 'The name(s) with largest drop in popularity between 1980 and 2014 was',' '.join(list(lowest_per_change.Name)) 
```

```python
# how have name lengths changed over time?
df2 = df.copy()
df2 = df2.groupby(['BirthYear','Name']).Count.sum().reset_index()
df2['Length'] = df2.Name.map(lambda x: len(x))
df2['WeightedLength'] = df2['Length'] * df2['Count']
df2 = df2.groupby(['BirthYear'])['Count','WeightedLength'].sum().reset_index()
df2['AverageLength'] = df2['WeightedLength'] / df2['Count']
df2
```

```python
plt.plot(df2['BirthYear'],df2['AverageLength'])
#plt.legend(['Average Name Length'],loc='upper right')
plt.ylim([5.0,7.0])
plt.xlim([1900,2020])
plt.ylabel('Average Name Length')
plt.xlabel('Year')
plt.show()
```

```python
# which letter changed popularity the most over time?
df4 = df.copy()
df4['FirstLetter'] = df4['Name'].str[0]
df4 = df4[(df4.BirthYear == 1920) | (df4.BirthYear == 2010)].reset_index()
df4['Count_1920'] = df4[df4.BirthYear == 1920].Count
df4['Count_2010'] = df4[df4.BirthYear == 2010].Count
df4.fillna(0,inplace=1)
df4 = df4.groupby(['FirstLetter'])['Count_1920','Count_2010'].sum().reset_index()
df4['Count_1920'] = df4['Count_1920'] / df4.Count_1920.sum()
df4['Count_2010'] = df4['Count_2010'] / df4.Count_2010.sum()
df4['Change'] = df4.Count_2010/df4.Count_1920
df4.sort('Change',ascending=False,inplace=1)
print df4.head()
print df4.tail()

# need to make the count columns a percentage
```

```python
df5 = df.copy()
df5['FirstLetter'] = df5['Name'].str[0]
df5_1 = df5.groupby(['BirthYear']).Count.sum().reset_index()
df5_2 = df5[df5.FirstLetter == 'Z']
df5_2 = df5_2.groupby(['BirthYear']).Count.sum().reset_index()
df5_2.columns = ['BirthYear','Count_Z']
df5_2 = df5_2.drop('BirthYear',1)
df5 = pd.concat([df5_1,df5_2],axis=1)
df5['Z_percentage'] = df5['Count_Z'] / df5['Count']

#most popular Z-starting name
df5_3 = df.copy()
df5_3['FirstLetter'] = df5_3['Name'].str[0]
df5_3 = df5_3[df5_3.FirstLetter=='Z']
df5_3 = df5_3.groupby(['Name']).Count.sum().reset_index()
df5_3 = df5_3[df5_3.Count == df5_3.Count.max()]
print df5_3.Name.iloc[0]

plt.plot(df5['BirthYear'],df5['Z_percentage'],color='green')
#plt.ylim([5.0,7.0])
plt.xlim([1910,2014])
plt.ylabel('Proportion of names starting with Z')
plt.xlabel('Year')
plt.fill_between(df5['BirthYear'],0,df5['Z_percentage'],color='green')
plt.show()
```
