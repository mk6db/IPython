```python
# what's cooking kaggle competition
# https://www.kaggle.com/c/whats-cooking/data?sample_submission.csv.zip
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
from collections import Counter

%matplotlib inline
```

```python
# import training data
train_file = 'C:/Users/Michael/projects/kaggle/whats_cooking/train.json'
traindf = pd.read_json(train_file)
```

```python
# sample of data
traindf.head()
```

```python
# create list of ingredients
ingredients_list = []
for i in range(len(traindf)):
    for j in traindf.ingredients[i]:
        if j not in ingredients_list:
            ingredients_list.append(j)
ingredients_list
```

```python
# add columns to df
traindf2 = pd.DataFrame(index=traindf.index, columns=ingredients_list)
traindf2.fillna(0)
traindf2.head()
```
