```python
# what's cooking kaggle competition
# https://www.kaggle.com/c/whats-cooking/data?sample_submission.csv.zip
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
from collections import Counter

%matplotlib inline
```

```python
# import training data
train_file = 'C:/Users/Michael/projects/kaggle/whats_cooking/train.json'
traindf = pd.read_json(train_file)
```

```python
# sample of data
traindf.head()
```

```python
# types of cuisines
cuisine_count = traindf.groupby('cuisine').count()
cuisine_count
```

```python
# create dictionary with all of the countries and counts of all food
bag_of_words = {}
for i in cuisine_count.index:
    cuisinedf = traindf[traindf.cuisine == i]
    bag_of_words[i] = {} 
    bag_of_words[i] = [Counter(recipe) for recipe in cuisinedf.ingredients]
sum_bags = {}
for i in bag_of_words:
    sum_bags[i] = {}
    sum_bags[i] = sum(bag_of_words[i],Counter())
sum_bags
```

```python
ingredients_df = pd.DataFrame(sum_bags).fillna(0)
ingredients_df = ingredients_df.transpose()
ingredients_df.head()
```

```python
# need to normalize data
# need to create a percentage for each column
ingredients_df['total'] = ingredients_df.sum(axis=1)
ingredients_df = ingredients_df.div(ingredients_df['total'], axis = 'index')
ingredients_df.head()
#sum_of_cols = []
#for i in ingredients_df.columns:
    #sum_of_cols.append(sum(ingredients_df[i]))
#print sum_of_cols
```

```python
#ingredients_df = ingredients_df.div(sum_of_cols, axis = 'columns')
ingredients_df = ingredients_df.drop('total',1)
ingredients_df.head()
```

```python
# import the class you plan to use
from sklearn.neighbors import KNeighborsClassifier
```

```python
# logreg imports
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from math import exp
```

```python
# create X and y
X = ingredients_df.values
y = ingredients_df.index
list(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1)
```

```python
# fit model
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
```

```python
print metrics.accuracy_score(y_test,y_pred)
```

```python
# now we have a weighted average dataframe
proj = []
for i in range(len(testdf.ingredients[0:500])):
    best = 'italian'
    previous_sum = 0
    for k in ingredients_df.columns[:-1]: 
        current_sum = 0
        for j in range(len(testdf.ingredients[i])):        
            if testdf.ingredients[i][j] in ingredients_df.index:
                current_sum += ingredients_df.ix[j,k]
        if current_sum > previous_sum:
            previous_sum = current_sum
            best = k
    proj.append(best)
proj
```
