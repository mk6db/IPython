```python
# create a stream of tweets
# this will append a list of tweets until the connection is lost

python_tweets = []

#override tweepy.StreamListener to add logic to on_status
class MyStreamListener(tweepy.StreamListener):

    def on_status(self, status):
        python_tweets.append(status.text)

    # adding on error handling    
    def on_error(self, status_code):
        if status_code == 420:
            #returning False in on_data disconnects the stream
            return False

# create a stream
myStreamListener = MyStreamListener()
myStream = tweepy.Stream(auth = api.auth, listener=myStreamListener)

# initiate a stream
myStream.filter(track=['python'])

# Tweepy offers a convenient async parameter on filter so the stream will run on a new thread. 
# myStream.filter(track=['python'], async=True)
```

```python
# after streaming, take a list of python related tweets and put them in a file
f = open('C:\home\output.txt','w')
for tweet in python_tweets:
    f.write("%s\n" % tweet.encode('utf-8'))
f.close()
```
