```python
# create a function that allows the user to search for a topic and return a list headlines
def headlines_NYT(query,begin_date='',end_date=''):
    '''This function will return headlines from the NY Times for the query term & optional news day input'''
    base_url = 'http://api.nytimes.com/svc/search/v2/articlesearch.json?q='
    query_input = ('+').join(query.split()) # adjust input to be used in url
    key = '&api-key=d24246bb069ae8de9408220269a8a7f1:13:73896407'
    begin_date_input = ''
    if begin_date:
        begin_date_input = '&begin_date=' + begin_date
    end_date_input = ''
    if end_date:
        end_date_input = '&end_date=' + end_date
    r = requests.get(base_url + query_input + begin_date_input + end_date_input + key) # request url
    articles = r.json() # convert out to json
    headlines = []
    for i in range(len(articles['response']['docs'])):
        headlines.append(articles['response']['docs'][i]['headline']['main'].encode('ascii','ignore'))
    return headlines
    

print headlines_NYT('babies')
print headlines_NYT('little one','20120530')
print headlines_NYT('sports','20120530','20151031')
```

```python
# create a function to collect abstracts
from collections import Counter

def home_abstracts_NYT():
    '''This function collects abstracts in the home section and returns a list'''
    abstracts = []
    url = 'http://api.nytimes.com/svc/topstories/v1/home.json?api-key=61ce730542e5d8de8adcd58950153c3a:5:73896407'
    r = requests.get(url)
    data = r.json()
    for article in range(len(data['results'])):
        abstracts.append(data['results'][article]['abstract'])
    return abstracts

home_abstracts_NYT()
```

```python
# create a function to product a table of the most popular words in the abstracts of top stories

from prettytable import PrettyTable
from collections import Counter

def pop_words_abstracts(list_of_abstracts):
    '''This function will return a table of the most popular words in the abstracts of top stories'''
    words = [w for a in list_of_abstracts for w in a.split()]
    pt = PrettyTable(field_names=['Word','Count'])
    c = Counter(words)
    [pt.add_row(kv) for kv in c.most_common()[:20]]
    pt.align['Word'], pt.align['Count'] = 'l','r'
    return pt

abstracts = home_abstracts_NYT()
print pop_words_abstracts(abstracts)
```
