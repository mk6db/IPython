```python
# once you have the forecast, you can look up different outputs
Currently = forecast.currently()
print Currently.summary
print Currently.icon
print

# by hour
byHour = forecast.hourly()
print byHour.summary
print byHour.icon
print

# can loop through hourly date (will output 48 hours)
for hourlyData in byHour.data:
    print hourlyData.time, hourlyData.d, hourlyData.temperature
    
# can load hourly data in dataframe
import datetime
import pandas as pd
weather = {'ozone':None,'temperature':None,'icon':None,'visibility':None,
          'summary':None,'apparentTemperature':None,'pressure':None,
          'windSpeed':None,'cloudCover':None,'time':None,'windBearing':None,
          'precipIntensity':None,'precipProbability':None}
weather_ = []
for hourlyData in byHour.data:
    weather['ozone'] = hourlyData.d['ozone']
    weather['temperature'] = hourlyData.d['temperature']
    weather['icon'] = hourlyData.d['icon']
    try:
        weather['visibility'] = hourlyData.d['visibility']
    except:
        weather['visibility'] = None
    weather['summary'] = hourlyData.d['summary']
    weather['apparentTemperature'] = hourlyData.d['apparentTemperature']
    weather['pressure'] = hourlyData.d['pressure']
    weather['windSpeed'] = hourlyData.d['windSpeed']
    weather['cloudCover'] = hourlyData.d['cloudCover']
    weather['time'] = datetime.datetime.fromtimestamp(hourlyData.d['time']).strftime('%Y-%m-%d %H:%M:%S') 
    weather['windBearing'] = hourlyData.d['windBearing']
    weather['precipIntensity'] = hourlyData.d['precipIntensity']
    weather['precipProbability'] = hourlyData.d['precipProbability']
    weather_.append(weather.copy())

cols = ['ozone','temperature','icon','visibility','summary','apparentTemperature','pressure','windSpeed','cloudCover','time','windBearing','precipIntensity','precipProbability']
weather_df = pd.DataFrame(weather_, columns = cols)
weather_df
```

```python
# create a function where the users inputs zip code and the hourly 
# weather forecast is output in a dataframe
import forecastio
import re
from geopy.geocoders import Nominatim

def hourly_weather():
    geolocator = Nominatim()
    
    zip = input("Enter your zip: ")
    zip = str(zip)

    zipCode = re.compile(r"^\d{5}(-\d{4})?$")

    if zipCode.match(zip):
        location = geolocator.geocode(zip)

        forecast = forecastio.load_forecast(api_key, location.latitude, location.longitude)

        byHour = forecast.hourly()

        weather = {'ozone':None,'temperature':None,'icon':None,'visibility':None,
          'summary':None,'apparentTemperature':None,'pressure':None,
          'windSpeed':None,'cloudCover':None,'time':None,'windBearing':None,
          'precipIntensity':None,'precipProbability':None}
        weather_ = []
        
        for hourlyData in byHour.data:
            weather['ozone'] = hourlyData.d['ozone']
            weather['temperature'] = hourlyData.d['temperature']
            weather['icon'] = hourlyData.d['icon']
            try:
                weather['visibility'] = hourlyData.d['visibility']
            except:
                weather['visibility'] = None
            weather['summary'] = hourlyData.d['summary']
            weather['apparentTemperature'] = hourlyData.d['apparentTemperature']
            weather['pressure'] = hourlyData.d['pressure']
            weather['windSpeed'] = hourlyData.d['windSpeed']
            weather['cloudCover'] = hourlyData.d['cloudCover']
            weather['time'] = datetime.datetime.fromtimestamp(hourlyData.d['time']).strftime('%Y-%m-%d %H:%M:%S') 
            weather['windBearing'] = hourlyData.d['windBearing']
            weather['precipIntensity'] = hourlyData.d['precipIntensity']
            weather['precipProbability'] = hourlyData.d['precipProbability']
            weather_.append(weather.copy())

        cols = ['ozone','temperature','icon','visibility','summary','apparentTemperature','pressure','windSpeed','cloudCover','time','windBearing','precipIntensity','precipProbability']
        weather_df = pd.DataFrame(weather_, columns = cols)
        return weather_df
    else:
        return "The zip is not valid."

hourly_weather()
```
