```python
# looking up fat tire we get a list of 50 beers
len(beers) #50

# we can create a pandas dataframe to store the results
import pandas as pd

beer = {'abv':None,'brewery':None,'brewery_id':None,'description':None,'ibu':None,\
        'id':None,'name':None,'organic':None,'style':None,'year':None}

beer_data = []
for i in range(len(beers)):
    beer['abv'] = beers[i].abv
    beer['brewery'] = beers[i].brewery
    beer['brewery_id'] = beers[i].brewery_id
    beer['description'] = beers[i].description
    beer['ibu'] = beers[i].ibu
    beer['id'] = beers[i].id
    beer['name'] = beers[i].name
    beer['organic'] = beers[i].organic
    beer['style'] = beers[i].style
    beer['year'] = beers[i].year
    beer_data.append(beer.copy())

cols = ['abv','brewery','brewery_id','description','ibu','id','name','organic','style','year']
beer_df = pd.DataFrame(beer_data, columns = cols)
beer_df
```

```python
# let's define a function to look up the beers and load them to dataframe
def beer_search(query):
    '''This function will return a pandas dataframe with details on beers that match the search term'''
    
    try:
        beers = brew_api.search_beer(query)
        beer = {'abv':None,'brewery':None,'brewery_id':None,'description':None,'ibu':None,\
            'id':None,'name':None,'organic':None,'style':None,'year':None}
        beer_data = []

        for i in range(len(beers)):
            beer['abv'] = beers[i].abv
            beer['brewery'] = beers[i].brewery
            beer['brewery_id'] = beers[i].brewery_id
            beer['description'] = beers[i].description
            beer['ibu'] = beers[i].ibu
            beer['id'] = beers[i].id
            beer['name'] = beers[i].name
            beer['organic'] = beers[i].organic
            beer['style'] = beers[i].style
            beer['year'] = beers[i].year
            beer_data.append(beer.copy())

        cols = ['abv','brewery','brewery_id','description','ibu','id','name','organic','style','year']
        beer_df = pd.DataFrame(beer_data, columns = cols)
        return beer_df
    except:
        return 'no beers found'

beer_search('bud')
```
